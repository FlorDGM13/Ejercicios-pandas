##CLASE EXTRA 05: CREANDO RANGOS DE VALOR

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler.csv', sep=';')
datos.head(10)
```

```python
#agrupar por la cantidad de cuartos
#1 y 2 cuarto
# 3 y 4 cuartos
# 5 y 6 cuartos
# 7 o mas cuartos

clases = [0, 2, 4, 6, 100]
```

```python
cuartos = pd.cut(datos.Cuartos, clases)
cuartos
```

```python
pd.value_counts(cuartos)
```

```python
label = ['1 y 2 cuartos', '3 y 4 cuartos', '5 y 6 cuartos', '7 cuartos o mas']
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label)
cuartos
```

```python
pd.value_counts(cuartos)
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label, include_lowest = True)
```

```python
pd.value_counts(cuartos)
```
